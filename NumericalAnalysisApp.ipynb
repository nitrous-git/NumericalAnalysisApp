## Lagrange Interpolation

Given $n+1$ distinct points $(x_i, f_i)$ with $f_i = f(x_i)$, we construct a polynomial $P_n(x)$ of degree at most $n$ such that $P_n(x_i) = f_i$ for $i = 0, \dots, n$.

### Lagrange Basis Polynomials

$$
\pi_i(x) = \prod_{\substack{j=0 \\ j \neq i}}^n \frac{x - x_j}{x_i - x_j}
$$

These satisfy $\pi_i(x_j) = 1$ if $i = j$, and $0$ otherwise.

### Interpolation Formula

$$
P_n(x) = \sum_{i=0}^n f_i \, \pi_i(x) = \sum_{i=0}^n f_i \prod_{\substack{j=0 \\ j \neq i}}^n \frac{x - x_j}{x_i - x_j}
$$

### Requirements

All abscissas must be distinct: $x_i \neq x_j$ for $i \neq j$.

### Complexity

Evaluating $P_n(x)$ costs $O(n^2)$ operations.

### Pros and Cons

- **Advantages**: Direct formula, theoretically clear, easy to implement.
- **Disadvantages**: Expensive for many evaluation points, impractical when adding points, can be numerically unstable for large $n$.

## Newton–Gregory Interpolation

The algorithm first constructs the forward difference table $\Delta^k f_i$, then evaluates the Newton–Gregory forward polynomial using the formula:

$$
P_n(x) = \sum_{k=0}^{n} \binom{s}{k} \, \Delta^k f_0, \quad s = \frac{x - x_0}{h}.
$$

Here, $h$ is the step between points (assumed to be equally spaced), and $\binom{s}{k}$ denotes the generalized binomial coefficient:

$$
\binom{s}{k} = \frac{s(s-1)(s-2)\cdots(s-k+1)}{k!}.
$$

The idea is simple: instead of constructing the polynomial directly in expanded form, we exploit the finite differences of the tabulated values. This approach is particularly well-suited when the abscissas are evenly spaced.

### Complexity

Constructing the difference table requires $O(n^2)$ operations, as it involves filling a triangular structure of quadratic size. Once the table is built, evaluating the polynomial at a given point $x$ costs only $O(n)$, since we sum $n+1$ terms.

### Error Estimation

The error can be estimated using the next term in the formula:

$$
E_n(x) \approx \binom{s}{n+1} \, \Delta^{n+1} f_0.
$$

This estimate requires at least one additional point beyond those used to construct $P_n$, because computing $\Delta^{n+1} f_0$ is necessary. In other words, a polynomial of degree $n$ is constructed with $n+1$ points, but its error is estimated using $n+2$ points.
